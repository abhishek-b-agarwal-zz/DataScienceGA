```python
def bed_colors(beds):
    if beds==1: return 'red'
    if beds==2: return 'green'
    if beds==3: return 'blue'
    return 'pink'
```

```python
sacramento.plot.scatter('longitude', 'latitude',c=sacramento.beds.map(bed_colors))
```

```python
sacramento.type.unique()
```

```python
sacramento['sale_day_of_year'] = sacramento.clean_date.dt.dayofyear
```

```python
residential = sacramento[sacramento.type == 'Residential']
condo = sacramento[sacramento.type == 'Condo']
```

```python
(fig,ax) = matplotlib.pyplot.subplots()
ax.plot(residential.groupby('sale_day_of_year').price.max(), c='green', label='Residential',marker='o')
ax.plot(condo.groupby('sale_day_of_year').price.max(), c='magenta', label='Condo', marker='^',linestyle='--')
ax.set_xlabel("Day of the year")
ax.set_ylabel("Price in ($)")
ax.legend(loc='best')
```
